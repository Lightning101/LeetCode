### Comments
***This problem can be solved using a lookup table.***

* The key lies in adding an entry for the in between values or subtractive values
### Complexity:

***Time***:O(n)

* Worst case we use ones for everything.

***Space***: O(n)

* Worst answer all ones and since the lookup table is fixed its constant.

````python
class Solution:
    def intToRoman(self, num: int) -> str:
        lookup = {
            1: "I",
            4: "IV",
            5: "V",
            9: "IX",
            10: "X",
            40: "XL",
            50: "L",
            90: "XC",
            100: "C",
            400: "CD",
            500: "D",
            900: "CM",
            1000: "M",
        }

        ans = []
        values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
        base = 0
        while num > 0:
            if num - values[base] >= 0:
                ans.append(lookup[values[base]])
                num -= values[base]
            else:
                base += 1

        return "".join(ans)
````

```
class Solution:
    def intToRoman(self, num: int) -> str:
        thousands = ["", "M", "MM", "MMM"]
        hundreds = ["", "C", "CC", "CCC", "CD", "D", "DC", "DCC", "DCCC", "CM"]
        tens = ["", "X", "XX", "XXX", "XL", "L", "LX", "LXX", "LXXX", "XC"]
        ones = ["", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX"]
        return (
            thousands[num // 1000]
            + hundreds[num % 1000 // 100]
            + tens[num % 100 // 10]
            + ones[num % 10]
        )

```

* Here we set out every posible combination and resolve the result.
