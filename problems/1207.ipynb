### Comments
***This problem can be solved using hashmap and a set***

### Complexity:

* Where K is the number of unique elements

***Time***: O(n + k)

***Space***: O(k)
* More realistically O(2k) as we keep a hashmap with all unique elements and then a set also for the count which if true becomes same length.

```python
from collections import defaultdict

class Solution:
    def uniqueOccurrences(self, arr: list[int]) -> bool:
        ans = defaultdict(int)

        for i in arr:
            ans[i] += 1

        check = set()

        for v in ans.values():
            if v in check:
                return False
            check.add(v)

        return True


s = Solution()


# Example 1:

# Input: arr = [1,2,2,1,1,3]
# Output: true
# Explanation: The value 1 has 3 occurrences, 2 has 2 and 3 has 1. No two values have the same number of occurrences.
print(s.uniqueOccurrences([1, 2, 2, 1, 1, 3]))

# Example 2:

# Input: arr = [1,2]
# Output: false
# Example 3:
print(s.uniqueOccurrences([1, 2]))

# Input: arr = [-3,0,1,-3,1,1,1,-3,10,0]
# Output: true

print(s.uniqueOccurrences([-3, 0, 1, -3, 1, 1, 1, -3, 10, 0]))
```
