### Comments
***This problem can be solved by using DFS with a hashmap***

* The key to this problem lies in traversing the graph and making copies of each node.
* We also need to keep track of each node we visited and since the edges are undirected for both nodes in an edge we need to do the connection.
* Hence why we use a map instead of a set.

### Complexity:

***Time***:O(n * e)

* This will traverse over every node of the graph and each connected edge.

***Space***: O(n)

* Where n is the number of nodes and we only use the map to keep a reference


```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []
"""

from typing import Optional


class Solution:
    def cloneGraph(self, node: Optional["Node"], visited: dict=None) -> Optional["Node"]:
        if not node:
            return None

        if not visited:
            visited = {}

        clonned = Node(node.val)
        visited[node.val] = clonned

        for neighbor in node.neighbors:
            if neighbor.val in visited:
                clonned.neighbors.append(visited[neighbor.val])
            else:
                clonnedNeighbor = self.cloneGraph(neighbor, visited)
                if clonnedNeighbor:
                    clonned.neighbors.append(clonnedNeighbor)

        return clonned
```
