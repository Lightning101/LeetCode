### Comments
***This problem can be solved using DFS***

* We must traverse down the tree and keep count if we are on a zig zag path.
* Otherise we must start counting again from the current node.

### Complexity:

***Time***:O(n)

***Space***: O(1)

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:

    def DFS(self, node, leftright, count):

        if not node:
            return count
        leftright = not leftright
        count += 1

        l = self.DFS(node.left, leftright, count if not leftright else 0)
        r = self.DFS(node.right, leftright, count if leftright else 0)
        return l if l > r else r

    def longestZigZag(self, root) -> int:

        l = self.DFS(root, False, 0)
        r = self.DFS(root, True, 0)
       
        return  (l if l > r else r) - 1
````

* Better solution that tarverses the tree only once.
```
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:

    def DFS(self, node, goLeft, count):

        if not node:
            return count

        count += 1
        l = self.DFS(node.left, False, count if goLeft else 1)
        r = self.DFS(node.right, True, count if not goLeft else 1)
        return l if l > r else r

    def longestZigZag(self, root: Optional[TreeNode]) -> int:

        return self.DFS(root, True, 0) - 1
```

* Here rather than trying to keep the leftright context we hard code it.
* This makes it so it does not matter where you start from and the context of which you start.
