***This problem can be solved using DFS***

* Base condition being that the node is valid.
* With us incrementing based on node.val >= currMax.

### Complexity:

***Time***:O(n)

***Space***: O(h)

* Where h is the height of the tree.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:

    def DFS(self,node,curr):
        if(not node):
            return 0

        if(node.val >= curr):
            curr = node.val
            return 1 +  self.DFS(node.left,node.val) + self.DFS(node.right,node.val)

        return self.DFS(node.left,curr) + self.DFS(node.right,curr)
        


    def goodNodes(self, root) -> int:
        return self.DFS(root,root.val)
```
