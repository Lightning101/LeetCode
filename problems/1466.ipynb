### Comments
***This problem can be solved with a DFS approach***

The keys to this problem lies in:

* Creating a adjacency list that is undirected whilst keeping the directionality hence we keep to values (connected_node, original_direction_of_connection).
* Using DFS to go over the graph like it was undirected.
* Since the graph is direct we could get loops so we keep a set to make sure we dont loop.

### Complexity:

***Time***:O(n)

***Space***: O(n)

````python
from collections import deque


class Solution:

    def minReorder(self, n: int, connections: list[list[int]]) -> int:

        adj = [[] for _ in range(n)]
        for i in range(len(connections)):
            s, r = connections[i]
            adj[s].append((r, 1))
            adj[r].append((s, 0))

        st = deque()
        st.append(0)
        count = 0
        visisted = set()
        visisted.add(0)
        while st:
            s = st.popleft()
            for i in range(len(adj[s])):
                d, c = adj[s][i]
                if d not in visisted:
                    count += int(c == 1)
                    st.append(d)
                    visisted.add(d)
        return count
````

### Community References:


* Given the fact that there is only n - 1 roads and all are connected there can be no cycles.
* As such we can remove the set.
* Pass along the parent and check the connection direction.
* Adding to the stack only when not the parent of the current node.

```
from collections import deque


class Solution:

    def minReorder(self, n: int, connections: list[list[int]]) -> int:

        adj = [[] for _ in range(n)]
        for i in range(len(connections)):
            s, r = connections[i]
            adj[s].append((r, 1))
            adj[r].append((s, 0))

        st = deque()
        st.append((0, -1))
        count = 0

        while st:
            c, p = st.popleft()
            for i in range(len(adj[c])):

                d, v = adj[c][i]
                if d != p:
                    count += v
                    st.append((d, c))
        return count

```
