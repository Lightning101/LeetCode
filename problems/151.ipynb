Problem can be solved by iterating backwards through the array after split opperation.


### Complexity:

***Time***: O(n)

***Space***: O(n)

```python
class Solution:
    def reverseWords(self, s: str) -> str:
        words = s.split(" ")
        ans = ""

        for i in range(len(words) - 1, -1, -1):
            if words[i] != "":
                ans += f"{words[i]} "
        
        ans = ans[:-1]

        return ans


s = Solution()
```
