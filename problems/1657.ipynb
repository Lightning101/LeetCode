### Comments
***This problem can be solved using a dictionary***

The best way is to adress the cases:

1. Same letters different order.
2. Different letters and different order.


* By using a dict we can solve the first on easy by just checking if the dicts are equal as they will have the same entries.
* Second case has to do with the count of each element if the count is the same for both dictionaries but the letter is different we can swap them.
* Laslty we need to check if the desired element exsits in the original.

eg:
```
! Same letters different order. 
"abc" -> "bca"

{a:1,b:1,c:1} == {b:1,c:1,a:1}


! Different letters and different order.
"cabbba" -> "abbccc"

{c:1,a:2,b:3} -> {a:1, b:2, c:3}

! When placed under each other we see we can check if a swap is possible
{c:1,a:2,b:3}
{a:1,b:2,c:3}

Which brings us back to the same case as Same letters different order

```

Since the element counts match up we can produce the desired output with the catch that the same elements exist in both.




### Complexity:

***Time***:O(n)

* You would think its O(n log(n)) due to the sorting but since there are only lowercase letters and 26 of them it becomes a O(26)

= O(n +26  + 26) = O(n) 


***Space***: O(n)

* Again here since the number of leters is fixed for our map

````python
from collections import Counter


class Solution:
    def closeStrings(self, word1: str, word2: str) -> bool:

        size1 = len(word1)
        size2 = len(word2)

        if size1 != size2:
            return False

        w1 = Counter(word1)
        w2 = Counter(word2)

        if w1 == w2:
            return True

        if w1.keys() != w2.keys():
            return False

        return sorted(w1.values()) == sorted(w2.values())


s = Solution()

# Example 1:

# Input: word1 = "abc", word2 = "bca"
# Output: true
# Explanation: You can attain word2 from word1 in 2 operations.
# Apply Operation 1: "abc" -> "acb"
# Apply Operation 1: "acb" -> "bca"
print(s.closeStrings(word1="abc", word2="bca"))
# Example 2:

# Input: word1 = "a", word2 = "aa"
# Output: false
# Explanation: It is impossible to attain word2 from word1, or vice versa, in any number of operations.

print(s.closeStrings(word1="a", word2="aa"))
# Example 3:

# Input: word1 = "cabbba", word2 = "abbccc"
# Output: true
# Explanation: You can attain word2 from word1 in 3 operations.
# Apply Operation 1: "cabbba" -> "caabbb"
# Apply Operation 2: "caabbb" -> "baaccc"
# Apply Operation 2: "baaccc" -> "abbccc"
print(s.closeStrings(word1="cabbba", word2="abbccc"))

# true
print(s.closeStrings(word1="abbzccca", word2="babzzczc"))


print(s.closeStrings(word1="uau", word2="ssx"))
````
