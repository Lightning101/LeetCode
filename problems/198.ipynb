* As can be seen by the bottom equation only the last 2 answers are relevant to calucate the current answer. As such we keep an array of 2 or in this case a queue and kick out the oldest element as we go.

### Complexity:

***Time***: O(n)

***Space***: O(1)

We keep an array of 2 elements and pick the largest as our answer.


````python
from collections import deque
class Solution:
    def rob(self, nums: list[int]) -> int:
        if len(nums) < 3:
            return max(nums)

        dp = deque([nums[0],max(nums[0], nums[1])])

        for i in range(2,len(nums)):
            temp = max([nums[i] + dp[-2], dp[-1]])
            dp.append(temp)
            dp.popleft()

        return max(dp)


s = Solution()

# Example 1:

# Input: nums = [1,2,3,1]
# Output: 4
# Explanation: Rob house 1 (money = 1) and then rob house 3 (money = 3).
# Total amount you can rob = 1 + 3 = 4.
print(s.rob([1,2,3,1]))
# Example 2:

# Input: nums = [2,7,9,3,1]
# Output: 12
# Explanation: Rob house 1 (money = 2), rob house 3 (money = 9) and rob house 5 (money = 1).
# Total amount you can rob = 2 + 9 + 1 = 12.
print(s.rob([2,7,9,3,1]))
````

class Solution:
    def rob(self, nums: list[int]) -> int:

        p2, p1 = 0, 0
        for i in range(len(nums)):
            curr = max([nums[i] + p2, p1])
            p2 = p1
            p1 = curr

        return max(p1,p2)
