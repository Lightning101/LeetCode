### Comments
***This problem can be solved using traversal***

* The key lies in iterating over both linked lists and adding whilst adding the carry over.
* To keep the answer we also create a empty head.

### Complexity:

***Time***: O(n)

* With n being the largest of the two linked lists.

***Space***: O(n)
* Where n is the number of digits in the linked list +2  for the final carry over and empty head.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode):

        ans = ListNode()
        carry = 0
        curr = ans
        while l1 and l2:
            v1 = l1.val
            v2 = l2.val
            total = v1 + v2 + carry

            curr.next = ListNode(total % 10)
            curr = curr.next

            carry = (total > 9) * 1
            l1 = l1.next
            l2 = l2.next


        while l1:
            v1 = l1.val
            total = v1 + carry

            curr.next = ListNode(total % 10)
            curr = curr.next

            carry = (total > 9) * 1
            l1 = l1.next

        while l2:
            v1 = l2.val
            total = v1 + carry

            curr.next = ListNode(total % 10)
            curr = curr.next

            carry = (total > 9) * 1
            l2 = l2.next

        if carry:
            curr.next = ListNode(1)
            curr = curr.next

        return ans.next
```
