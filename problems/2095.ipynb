### Comments
***This problem can be solved by using 2 pointer approach***

Commonly refered to as the fast and slow pointer.
* Here we initialize 2 pointers and increment then by 1 and 2 respectively.
* Once the fast pointer becomes null we have the middle point of the list.
* Small change here is that a prev variable is kept to store the element before the middle of the swap. Helps avoid having to iterate again.

* Edge case is when there is only one element then prev == p1 and we should return None.


### Complexity:

***Time***:O(n)

***Space***: O(1)

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteMiddle(self, head):
        p1,p2 = head,head
        prev = head
        while p2 and p2.next:
            prev = p1
            p1 = p1.next
            p2 = p2.next.next

        if(prev == p1):
            return None

        prev.next = p1.next
        return head
````

```
class Solution:
    def deleteMiddle(self, head: Optional[ListNode]) -> Optional[ListNode]:

        if(head.next == None):
            return None

        p1,p2 = head,head.next.next

        while p2 and p2.next:
            p1 = p1.next
            p2 = p2.next.next


        p1.next = p1.next.next
        return head
```

* The edge case of checking if only 1 element in list is done initially.
* This version also stops 1 before the last element removing the need for keeping prev.
* This is done my intializing the p2 to head.next.next initially.
