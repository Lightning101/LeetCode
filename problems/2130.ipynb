***This problem can be solved using the fast slow pointer approach***

* This problem has 2 keys to it:
* * Since evey twin is exactly n- i-1 which is the reversed of the list from the middle.
* * After which we can just do normal iteration and calculate the max twins.

* By using the common fast and slow pointer approach to find the middle of the list. 
* We also reverse the list as we go.  

### Complexity:

***Time***:O(n)

***Space***: O(1)

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def pairSum(self, head) -> int:
        p1, p2 =  head, head.next
        prev = None

        while(p2 and p2.next):
            p2= p2.next.next

            temp = p1.next
            p1.next = prev
            prev= p1
            p1 = temp


        p2 = p1.next
        p1.next = prev
        

        maxSum = 0
        while(p1 and p2 ):
            if((currSum:= p1.val + p2.val) > maxSum):
                maxSum = currSum
            p1 = p1.next
            p2 = p2.next

        return maxSum
            
```
