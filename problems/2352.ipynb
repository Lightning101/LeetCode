### Comments
***This problem can be solved using dicts***

* Idea is to keep count of how many unique rows and cols we come across.
* Then we multiply the pair togther.

### Complexity:

***Time***:O(n^2 + n)

***Space***:O(n)

```python
from collections import defaultdict


class Solution:
    def equalPairs(self, grid: list[list[int]]) -> int:

        size = len(grid)
        columns = defaultdict(int)
        rows = defaultdict(int)

        for x in range(size):
            column = []
            for y in range(size):
                column.append(grid[y][x])

            columns[tuple(column)] += 1
            rows[tuple(grid[x])] += 1
        ans = 0

        for k in columns.keys():
            if k in rows:
                ans += columns[k] * rows[k]

        return ans


s = Solution()


# Example 1:


# Input: grid = [[3,2,1],[1,7,6],[2,7,7]]
# Output: 1
# Explanation: There is 1 equal row and column pair:
# - (Row 2, Column 1): [2,7,7]

print(s.equalPairs([[3, 2, 1], [1, 7, 6], [2, 7, 7]]))
# Example 2:


# Input: grid = [[3,1,2,2],[1,4,4,5],[2,4,2,2],[2,4,2,2]]
# Output: 3
# Explanation: There are 3 equal row and column pairs:
# - (Row 0, Column 0): [3,1,2,2]
# - (Row 2, Column 2): [2,4,2,2]
# - (Row 3, Column 2): [2,4,2,2]

print(s.equalPairs([[3, 1, 2, 2], [1, 4, 4, 5], [2, 4, 2, 2], [2, 4, 2, 2]]))
```
