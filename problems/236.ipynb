We try and address the following cases:

* p and q are both child decendents of T.
* T is one of p or q.

### Complexity:

***Time***:O(n)

***Space***:O(1)

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


class Solution:

    def DFS(self, node, p, q):
        if not node:
            return None

        left = self.DFS(node.left, p, q)
        right = self.DFS(node.right, p, q)

        # Case 1,2
        if (
            node == p
            or node == q
            or (left == p and right == q)
            or (left == q and right == p)
        ):
            return node
        elif left != right:
            return left if left else right
        
        return None

    def lowestCommonAncestor(
        self, root , p , q 
    ):
        return self.DFS(root, p, q)
````

```
class Solution:

    def lowestCommonAncestor(
        self, node: "TreeNode", p: "TreeNode", q: "TreeNode"
    ) -> "TreeNode":
        if not node:
            return None

        left = self.lowestCommonAncestor(node.left, p, q)
        right = self.lowestCommonAncestor(node.right, p, q)

        # Case 1,2
        if node == p or node == q or (left and right):
            return node
        return left or right
```

* This version works on the assumption that the answer is always present.
