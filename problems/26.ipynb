***This problem can be solved using a 2 pointers approach***

* The idea is walk the line and keep track of valid positions.

* Here p is the location of the last unique element we have come across.

```
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        size = len(nums)

        p = 0
        for i in range(1, size):
            if nums[p] != nums[i]:
                nums[i], nums[p + 1] = nums[p + 1], nums[i]
                p += 1
        return p +1
```

* To make it clearer the above example is more better since i start p from 0. I changed it so we dont end up having to add one at the end in the case of boundry condition.


### Complexity:

***Time***:O(n)

***Space***: O(1)

````python
class Solution:
    def removeDuplicates(self, nums: list[int]) -> int:
        size = len(nums)
        p = 1
        for i in range(1, size):
            if nums[p - 1] != nums[i]:
                nums[i], nums[p] = nums[p], nums[i]
                p += 1
        return p
````
