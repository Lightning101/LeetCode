*** This problem can be solved using a 2 pointer approach ***

Following the conditions:

* If both p1 and p2 are not vowels iterate them towards the center.
* If both p1 and p2 are vowels swap them and then iterate.
* If only p1 is a vowel iterate p2.
* If only p2 is a vowel iterate p1.
* If p1 < p2 break the loop

### Complexity:

***Time***: O(n)

***Space***: O(n)
* Plus the vowels(10) we keep in a set for fast checks

````python
class Solution:
    def reverseVowels(self, s: str) -> str:
        ans = list(s)
        p1, p2 = 0, len(s) - 1

        vowels = set(["a", "e", "i", "o", "u", "A", "E", "I", "O", "U"])
        while p1 < p2:
            if ans[p1] in vowels and ans[p2] in vowels:
                ans[p1], ans[p2] = ans[p2], ans[p1]
                p1 += 1
                p2 -= 1
            elif ans[p1] in vowels:
                p2 -= 1
            elif ans[p2] in vowels:
                p1 += 1
            else:
                p1 += 1
                p2 -= 1
        return "".join(ans)

s = Solution()


print(s.reverseVowels("IceCreAm") == "AceCreIm")
print(s.reverseVowels("leetcode") == "leotcede")
````
