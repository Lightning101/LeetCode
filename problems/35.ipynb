### Comments
***This problem can be solved using binary search***

The problem is a standard binary search with the only variation being that we need to return the insert position if the element is not found.


### Complexity:

***Time***:O(log(n))

***Space***: O(1)

```python
class Solution:
    def searchInsert(self, nums: list[int], target: int) -> int:
        l, r = 0, len(nums) - 1

        while l <= r:
            mid = (l + r) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] > target:
                r = mid - 1
            else:
                l = mid + 1

        return l
```
