### Comments
*** This problem can be solved using sets ***

* The key lies in keeping track of the columns,rows and boxes.

### Complexity:

***Time***:O(1)

* More acuratley O(n) for the insertion into a set worst case and since its fixed bounds it becomes O(1).

***Space***: O(1)

* Since we remove the columns and rows and only keep the boxs and with the boxs having a fixed bounds of 9*9 we get O(81) ~ O(1)

```python
class Solution:
    def isValidSudoku(self, board: list[list[str]]) -> bool:
        boxs = [set() for _ in range(9)]

        for x in range(9):
            column = set()
            row = set()
            for y in range(9):
                rowVal = board[x][y]
                colVal = board[y][x]

                
                if rowVal != "." and (rowVal in boxs[y // 3 + 3 * (x // 3)] or  rowVal in row):
                    return False
                row.add(rowVal)
                boxs[y // 3 + 3 * (x // 3)].add(rowVal)
                
                if colVal != "." and colVal in column:
                    return False
                column.add(colVal)

                

        return True
```
