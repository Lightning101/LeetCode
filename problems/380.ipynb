* !Note that my solution is not really a valid solution as it is not truly random.

### Complexity:

***Time***:O(1)

* Average O(1) even though worst case can be O(n) for hashing the number

***Space***: O(n)

* With n being the number of inserts.

````python
import random


class RandomizedSet:

    def __init__(self, maxRand=10):
        self.data = set()
        self.maxRand = maxRand

    def insert(self, val: int) -> bool:
        exists = not val in self.data
        self.data.add(val)
        return exists

    def remove(self, val: int) -> bool:
        if val in self.data:
            self.data.remove(val)
            return True
        return False

    def getRandom(self) -> int:
        it = iter(self.data)
        idx = random.randrange(0, min(self.maxRand, len(self.data)))
        for _ in range(idx):
            next(it)
        return next(it)
````

* Far better approach with truely random is to use a map and a array. Taking advanatge of the fact that removal from the end of an array is O(1) by swapping the elements.

```

import random


class RandomizedSet:

    def __init__(self):
        self.keyMap = {}
        self.data = []

    def insert(self, val: int) -> bool:
        if val in self.data:
            return False
        self.keyMap[val] = len(self.data)
        self.data.append(val)

    def remove(self, val: int) -> bool:
        if val in self.keyMap:
            self.data[self.keyMap[val]], self.data[-1] = (
                self.data[-1],
                self.data[self.keyMap[val]],
            )
            self.keyMap[self.data[self.keyMap[val]]] = self.keyMap[val]
            del self.keyMap[val]
            return True
        return False

    def getRandom(self) -> int:
        return random.choice(self.data)
```
