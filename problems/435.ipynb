### Comments
***This problem can be solved by sorting***

The keys to this problem lie in the following cases:

* If the interval doesnt overlap at all: start[1] <= curr[0]
* If the current interval is larger than the start then remove the bigger one: start[1] < curr[1]
* If our starting interval is bigger or equal to the current interval: start[1] >= curr[1]

### Complexity:

***Time***:O(nlog(n))

* As we sort first before iterating

***Space***: O(1)

````python
class Solution:
    def eraseOverlapIntervals(self, intervals: list[list[int]]) -> int:
        intervals.sort()
        ans = 0
        start = intervals[0]
        
        for i in range(1, len(intervals)):
            curr = intervals[i]
            if  start[1] <= curr[0]:
                start = curr
            elif start[1] < curr[1]:
                ans += 1
            elif start[1] >= curr[1]:
                start = curr
                ans += 1
        return ans


s = Solution()


# Example 1:

# Input: intervals = [[1,2],[2,3],[3,4],[1,3]]
# Output: 1
# Explanation: [1,3] can be removed and the rest of the intervals are non-overlapping.
# [[1,2],[1,3],[2,3],[3,4]]
# print(s.eraseOverlapIntervals([[1, 2], [2, 3], [3, 4], [1, 3]]))
# Example 2:

# Input: intervals = [[1,2],[1,2],[1,2]]
# Output: 2
# Explanation: You need to remove two [1,2] to make the rest of the intervals non-overlapping.

# print(s.eraseOverlapIntervals([[1, 2], [1, 2], [1, 2]]))
# Example 3:

# Input: intervals = [[1,2],[2,3]]
# Output: 0
# Explanation: You don't need to remove any of the intervals since they're already non-overlapping.

# print(s.eraseOverlapIntervals([[1, 2], [2, 3]]))

# 2
print(s.eraseOverlapIntervals([[1,100],[11,22],[1,11],[2,12]]))
````
