### Comments
***This problem can be solved using index rules***

* The key lies in recording how many x^10, x^100 .... all the way to x^10e31
* This allows us to create a base of x^10... and multiply it can keep adding to the answer.

### Complexity:

***Time***:O(log10(n))

* Should be log10(n) but I find it takes longer especially with any x~>2 the sheer base multiplication slows down the computation.

***Space***: O(1)

```python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        powers = [0 for _ in range(31)]
        largest = 1
        temp = abs(n)
        for i in range(31):
            powers[i] = temp % 10
            if powers[i] > 0:
                largest = i + 1
            temp //= 10

        ans = 1
        for i in range(powers[0]):
            ans *= x

        base = x * x * x * x * x * x * x * x * x * x
        for i in range(1, largest):

            total = 1
            for _ in range(powers[i]):
                total *= base
            ans *= total
            base *= base * base * base * base * base * base * base * base * base

        if n < 0:
            return 1 / ans
        return ans
```
