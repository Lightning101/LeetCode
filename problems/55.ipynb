### Complexity:

***Time***:O(n)

***Space***: O(n)

````python
class Solution:
    def jump(self, i,nums, memo):
        if i in memo:
            return memo[i]
        if i >= len(nums) - 1:
            return True
        for j in range(nums[i], 0, -1):
            if self.jump(i + j,nums, memo):
                return True
        memo[i] = False
        return memo[i]

    def canJump(self, nums: list[int]) -> bool:
        if len(nums) <= 1:
            return True
        return self.jump(0,nums, {})


s = Solution()


print(s.canJump([2,0,6,9,8,4,5,0,8,9,1,2,9,6,8,8,0,6,3,1,2,2,1,2,6,5,3,1,2,2,6,4,2,4,3,0,0,0,3,8,2,4,0,1,2,0,1,4,6,5,8,0,7,9,3,4,6,6,5,8,9,3,4,3,7,0,4,9,0,9,8,4,3,0,7,7,1,9,1,9,4,9,0,1,9,5,7,7,1,5,8,2,8,2,6,8,2,2,7,5,1,7,9,6]))
````
