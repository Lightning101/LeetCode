### Comments
***This problem can be solved with one iteration by considering that a 1 can be either contructive or destructive(creating gap) for us.***

* We assume the start is a one and add that in as nessary.
* Then we check for 3 cases:
    * Either both i and i-1 equal 0 then we can add a 1 and increment the answer.
    * Either both i and i-1 equal 1 then we need to decrement the answer as the preplanted 1 causes a gap in our ideal scenario.
    * There are no preplanted plants so we can just return ***(size + 1) // 2 >= n***




### Complexity:

***Time***: O(n)

***Space***: O(1)


```python
class Solution:
    def canPlaceFlowers(self, fb: list[int], n: int) -> bool:
        size = len(fb)
        if any(fb):
            ans = 0 if fb[0] else 1
            fb[0] = 1
            for i in range(1, size):
                if fb[i - 1] == 0 and fb[i] == 0:
                    fb[i] = 1
                    ans += 1
                elif fb[i - 1] == 1 and fb[i] == 1:
                    fb[i - 1] = 0
                    ans -= 1
            return ans >= n
        else:
            return (size + 1) // 2 >= n


s = Solution()


# Example 1:

# Input: flowerbed = [1,0,0,0,1], n = 1
# Output: true
print(s.canPlaceFlowers([1, 0, 0, 0, 1], 1))
# Example 2:

# Input: flowerbed = [1,0,0,0,1], n = 2
# Output: false
print(s.canPlaceFlowers([1, 0, 0, 0, 1], 2))
```
