### Comments
***This problem can be solved using stacks***

* The approach is to split the path using '/' as the delemiter.
* Reverse the list to allow us to use the end for fast pop.
* Then we address each case induvidually.
  * '.' or empty string are skipped.
  * '..' is used to pop off our answer stack
  * Anything else is considered a valid directoy or file so we push to our ans

### Complexity:

***Time***: O(n) 

***Space***: O(n)

```python
class Solution:
    def simplifyPath(self, path: str) -> str:
        if path[0] != "/":
            return None
        ans = []
        st = path.split("/")[::-1]

        while st:
            curr = st.pop()
            if curr == "." or curr == "":
                continue
            elif curr == "..":
                if(ans):
                    ans.pop()
            else:
                ans.append(curr)

        return "/" + "/".join(ans)
```
