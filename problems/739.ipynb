A more intresting approach outlined by LeetCode was by using the information in the answers to basically jump to the next answer. This can be achieved by iterating backwards.

```
class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        n = len(temperatures)
        hottest = 0
        answer = [0] * n
        
        for curr_day in range(n - 1, -1, -1):
            current_temp = temperatures[curr_day]
            if current_temp >= hottest:
                hottest = current_temp
                continue
            
            days = 1
            while temperatures[curr_day + days] <= current_temp:
                # Use information from answer to search for the next warmer day
                days += answer[curr_day + days]
            answer[curr_day] = days

        return answer
```

### Complexity:

***Time***: O(n)

***Space***: O(n)

````python
from collections import deque
class Solution:
    def dailyTemperatures(self, temps: list[int]) -> list[int]:
        size = len(temps)
        ans = [0] * size

        st = deque()
        for i in range(size):
            while(st and temps[st[-1]] < temps[i]):
                idx = st.pop()
                ans[idx] = i - idx
            st.append(i)
        
        return ans


s = Solution()



# Example 1:

# Input: temperatures = [73,74,75,71,69,72,76,73]
# Output: [1,1,4,2,1,1,0,0]

print(s.dailyTemperatures([73,74,75,71,69,72,76,73]))
# Example 2:

# Input: temperatures = [30,40,50,60]
# Output: [1,1,1,0]
print(s.dailyTemperatures([30,40,50,60]))
# Example 3:

# Input: temperatures = [30,60,90]
# Output: [1,1,0]
print(s.dailyTemperatures([30,60,90]))
````
