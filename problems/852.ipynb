***This problem can be solved using binary search***

* The key lies in the condition that we use check if we are on the rising section or falling section:

```
if (mid + 1) < len(arr) and arr[mid] < arr[mid + 1]:
```

* If the last element is peak we need to make sure we dont go over it.

### Complexity:

***Time***: O(log(n))

***Space***: O(1)

````python
class Solution:
    def peakIndexInMountainArray(self, arr: list[int]) -> int:

        l, r = 0, len(arr)

        while l <= r:
            mid = (l + r) // 2
            if (mid + 1) < len(arr) and arr[mid] < arr[mid + 1]:
                l = mid + 1
            else:
                r = mid - 1

        return l
````
